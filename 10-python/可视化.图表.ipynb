## 二、折线图

````python
import numpy as np

np.random.seed(55)
df = pd.DataFrame({"宁德时代": np.random.randn(100)+0.2,
                   "贵州茅台": np.random.randn(100)+0.17},
                  index=pd.date_range('1/1/2021', periods=100))


# 默认行累加
df = df.cumsum()
print(df)
# 每个值加50
df = df + 50
print(df)
df.plot_bokeh(kind="line")
````
